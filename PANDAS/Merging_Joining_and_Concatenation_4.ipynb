```python
import numpy as np
import pandas as pd
```

Merging 2 Dataframes

```python
# DataFrame 1: Employee Info
employees = pd.DataFrame({
    'employee_id': [1,2,3,4,5],
    'name': ['John','Anna', 'Peter','Linda', 'Bob'],
    'dept': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary Info
salaries = pd.DataFrame({
    'employee_id': [1,2,3,6,7],
    'salary': [1000, 2300, 4300, 2200, 2700],
    'bonus': [400, 700, 1200, 600, 750]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries)
```

```python
pd.merge(employees,salaries, on='employee_id', how='inner')
```

```python
pd.merge(employees,salaries, on='employee_id', how='outer')
```

```python
pd.merge(employees,salaries, on='employee_id', how='left')
```

```python
pd.merge(employees,salaries, on='employee_id', how='right')
```

Concatenation of 2 DataFrames

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis = 1)
```

Joining 2 Data Frames

```python
# First Data Frame
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index = [1,2,3])

# Second Data Frame
df2 = pd.DataFrame({
    'score': [85,90,75]
}, index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```
