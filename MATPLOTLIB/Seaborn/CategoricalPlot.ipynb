# Boxplot and Violinplot  

**Boxplots** and **Violinplots** are used to show the distribution of categorical data.  

- A **box plot** (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.  
- The **box** shows the quartiles of the dataset, while the **whiskers** extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.  


```python
# Violinplot
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = df)
```

# Stripplot and Swarmplot  

**Stripplot**  
- A strip plot draws a scatterplot where one variable is categorical.  
- It can be drawn on its own, but is often used as a complement to a **boxplot** or **violinplot**.  
- Useful when you want to show **all observations** along with some summary of the underlying distribution.  

**Swarmplot**  
- A swarmplot is similar to a stripplot, but the points are adjusted (only along the categorical axis) so that they **don’t overlap**.  
- This gives a clearer representation of the distribution of values.  
- Limitation: Does not scale well to very large datasets (harder to plot and interpret when there are too many points).  


```python
# Stripplot
```

```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
```

```python
# Swarmplot
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.violinplot(x='tip', y='day', data=df)
sns.swarmplot(x='tip', y='day', data=df, size=3, color='red')
```
