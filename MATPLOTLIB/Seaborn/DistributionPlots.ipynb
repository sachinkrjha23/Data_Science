```python
import matplotlib.pyplot as plt
import seaborn as sns
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

• distplot/histplot 

• jointplot

• pairplot

• rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
# Distplot or Histplot
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True, color = 'red')
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
# Jointplot
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
```

```python
# Pairplot
```

```python
sns.pairplot(df,hue='size', palette='rainbow')
```
