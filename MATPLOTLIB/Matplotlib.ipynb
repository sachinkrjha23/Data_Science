```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
y = x **3 
y
```

```python
plt.plot(x,y)
plt.title("My first Matplotlib Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

Object Oriented Way

```python
fig = plt.figure(figsize=(20,8),dpi = 80)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

Types of Plots

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot_1.png")
```

Working with Images

```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('me.jpeg')

plt.imshow(img)
plt.axis('off')
```

```python
cropped_img = img[50:250, 100:300]  # crop: rows 50-250, cols 100-300
plt.imshow(cropped_img)
plt.axis('off')
```
